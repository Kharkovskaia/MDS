### Задание 1:
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, t
import math
```

```python
# Нормальное распределение
norm_items = [(0.5, 0.5), (1, 0.5), (0.5, 1), (1, 1), (1, 0.75)]
for item in norm_items:
    my_norm = norm(loc=item[0], scale=item[1])
    x = np.linspace(-2, 4, 100)
    y = my_norm.pdf(x)
    plt.plot(x, y, label=f"µ = {item[0]}, σ = {item[1]}")
plt.title("Нормальное рапределение")
plt.legend()
plt.show()
```

Из полученных значений видно, что µ определяет нахождение максимального значения плотности распределения и при изменении этого параметра пик плотности расределения смещаается по графику.
σ определяет высоту плотности вероятности. Чем больше σ, тем больше высота плотности распределения.

```python
# Экспоненциальное распределение
expon_items = [0, 0.5, 1, 1.5, 2]
for item in expon_items:
    my_expon = expon(item)
    x = np.linspace(-2, 6, 1000)
    y = my_expon.pdf(x)
    plt.plot(x, y, label=f"λ = {item}") 
plt.title("Экспоненциальное рапределение")
plt.legend()
plt.show()
```

Из полученных значений виднj, что чем больше λ, тем больше график смещен вправо.

```python
# Распределение Стьюдента
t_items = [0, 0.5, 1, 1.5, 2]
for item in t_items:
    my_t = t(item)
    x = np.linspace(-5, 5, 1000)
    y = my_t.pdf(x)
    plt.plot(x, y, label=f"n = {item}") 
plt.title("Распределение Стьюдента")
plt.legend()
plt.show()
```

Из полученных значений видно, что чем больше число степеней свбоды n, тем более высокий и узкий график плотности распределения.

### Задание 2
Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания в отрезок $[3.5;7]$, т.е. P(3.5 < $\xi$ < 7).

C = 1 / (b - a) = 1 / (5 - 0) = 0,2

Найдем математическое ожидание:


ξ = (a + b) / 2 = (0 + 5) / 2 = 2,5

Найдем вероятность попадания в отрезок [3.5;7]:

P(3.5 < ξ < 7) = P (3.5 < ξ <= 5) + P(5 < ξ < 7) = (5 - 3.5) * 1 / 5 + 0 = 0.3

### Задание 3
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
# Найдем какое количество информации несет сообщение о цвете светофора в текущий момент
# при том, что светофор горит 54 секунды красным, 4 секунды желтым и 28 секунд зеленым
r = 54 #Количество времени, когда светофор горит красным
y = 4 #Количество времени, когда светофор горит желтым
g = 28 #Количество времени, когда светофор горит зеленм
r_y_g = r + y + g
p_r = r / r_y_g
p_y = y / r_y_g
p_g = g / r_y_g
i_r = -math.log2(p_r)
i_y = -math.log2(p_y)
i_g = -math.log2(p_g)
print(f"Количество информации для красного цвета {round(i_r, 3)}")
print(f"Количество информации для желтого цвета {round(i_y, 3)}")
print(f"Количество информации для зеленого цвета {round(i_g, 3)}")
```

```python
# Аналогично найдем какое количество информации несет сообщение о цвете светофора в текущий момент
# в случае 28, 28 и 28 секунд для каждого цвета
r = 28 #Количество времени, когда светофор горит красным
y = 28 #Количество времени, когда светофор горит желтым
g = 28 #Количество времени, когда светофор горит зеленм
r_y_g = r + y + g
p_r = r / r_y_g
p_y = y / r_y_g
p_g = g / r_y_g
i_r = -math.log2(p_r)
i_y = -math.log2(p_y)
i_g = -math.log2(p_g)
print(f"Количество информации для красного цвета {round(i_r, 3)}")
print(f"Количество информации для желтого цвета {round(i_y, 3)}")
print(f"Количество информации для зеленого цвета {round(i_y, 3)}")
```

Из полученных значений видно, что чем больше горит цвет, тем меньше передается информации.
В случае, когда все цвета горят одинаковое время, количество информации, которое несет сообщение о цвете, не изменяется в зависимости от цвета.
