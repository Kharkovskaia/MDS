```python
from scipy.misc import derivative
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
```

## Уровень 0:
Посчитайте значение первой и второй производной функции:

В точке x = 1.

```python
def f_1(x):
    return x ** 5 + 4 * sin(2 * x) + cos(3 * x + 3)
```

```python
print('Значение первой производной функции в точке х = 1:', derivative(f_1, 1.0, dx=1e-6))
print('Значение первой производной функции в точке х = 1:', derivative(f_1, 1.0, dx=1e-6, n=2))
```

## Уровень 1:
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
x = symbols('x')
```

```python
# Вносим значение функции
def f_2(x):
    return((sin((2.0 * x) + 1.0)) ** 5.0)
```

```python
# Находим значения x на отрезке [-5; 5]
item_x = np.linspace(-5, 5, num=50, endpoint=True, retstep=False, dtype=None)
```

```python
# Строим график первой производной
plt.plot(item_x, [derivative(f_2, x, dx=1e-6) for x in item_x])
```

```python
# Строим график второй производной
plt.plot(item_x, [derivative(f_2, x, dx=1e-6, n=2) for x in item_x])
```
